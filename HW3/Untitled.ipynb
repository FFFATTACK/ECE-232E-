```r
rf1<-matrix(0L,nrow<-10,ncol<-10)
rf1[10,10]<-1.0
```

```r
rownames(rf1)<-c(0:9)
colnames(rf1)<-c(0:9)
heatmap(rf1,Rowv=NA, Colv=NA,scale="none",col=cm.colors(256))
```
